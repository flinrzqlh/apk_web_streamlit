# Predict Sentiment per Kalimat

```
import re

original_text =  '''
Masakan ini ngga enak, cuih.
'''

def cleansing(sent):
    # Mengubah kata menjadi huruf kecil semua dengan menggunakan fungsi lower()
    string = sent.lower()
    # Menghapus emoticon dan tanda baca menggunakan "RegEx" dengan script di bawah
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string)
    # Menghapus RT
    string = re.sub(r'rt', '', string)
    return string

# cleansing
text = cleansing(original_text)
```

```
# feature extraction
text_feature = feature_bow.transform([text])
```

```
# predict
print("Text: ", str(original_text))
print("Sentiment NB: ", model_nb.predict(text_feature)[0])
print("Sentiment NN: ", model_nn.predict(text_feature)[0])
```

# Predict Sentiment dari Kumpulan Kalimat

```
import pandas as pd

data = pd.read_csv('drive/My Drive/Python/NLP/2024-nlp/tapera/tapera.csv', sep=',')
```

```
data = data.rename(columns={'full_text': 'Tweet'})
```

```
data.head()
```

```
import re

def cleansing(sent):
    # Mengubah kata menjadi huruf kecil semua dengan menggunakan fungsi lower()
    string = sent.lower()
    # Menghapus emoticon dan tanda baca menggunakan "RegEx" dengan script di bawah
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string)
    # Menghapus RT
    string = re.sub(r'rt', '', string)
    return string
```

```
def predict_sentiment(sent):
    # cleansing
    text = cleansing(str(sent))
    # feature extraction
    text_feature = feature_bow.transform([text])
    # predict
    return model_nb.predict(text_feature)[0]
```

```
data['Tweet'] = data.Tweet.apply(cleansing)
data.head()
```

```
data['predicted_sentiment'] = data.Tweet.apply(predict_sentiment)
```
